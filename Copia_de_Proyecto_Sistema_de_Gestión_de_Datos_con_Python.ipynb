# **Tema: Se realiza un menú CRUD referente a animales (perros y gatos) disponibles para adopción**

```python
#Se define el catalógo de animales en adopción

catalogo_animales = [{'nombre':'Penny', 'especie':'gato', 'género':'hembra', 'edad en años': '2'},
                     {'nombre':'Bruno', 'especie':'perro', 'género':'macho', 'edad en años': '0.5'},
                     {'nombre':'Lola', 'especie':'perro', 'género':'hembra', 'edad en años': '5'},
                     {'nombre':'Lucas', 'especie':'perro', 'género':'macho', 'edad en años': '1'}]

catalogo_animales
```

```python
# Crea una función para el menú CRUD del catálogo

def menu_opciones():
  menu_opc='''
  *****************************************************************
  Menú CRUD - Catalógo de animales en adopción:
  *****************************************************************
  1. Ver catálogo
  2. Crear registro             _____
  3. Consultar registro       (| * * |)      ^___^
  4. Actualizar registro       ( (Y) )     (=' X '=)
  5. Eliminar registro          (")(")_/    (")_(")_/
  6. Salir del menú
  *****************************************************************
'''
  print(menu_opc)

menu_opciones()
```

```python
# SCrear función mostrar agenda

from ast import Index


def ver_catalogo(catalogo_animales):

  gatos = 0

  print("\n------------------------------------------")
  print('Animales disponibles para adopción: ')
  print("------------------------------------------")
  for animal in catalogo_animales:
    print(f"\nAnimal {catalogo_animales.index(animal)+1}:")
    print(f"nombre: {animal['nombre']}")
    print(f"especie: {animal['especie']}")
    print(f"género: {animal['género']}")
    print(f"edad en años: {animal['edad en años']}")
    print("------------------------------------------")

ver_catalogo(catalogo_animales)

```

```python
#Se crea la opción para crear un registro

def crear_registro():
  print("\n------------------------------------------")
  print("Ingresa los datos del registro a crear ")
  print("------------------------------------------\n")

  registro={'nombre': input('Ingresa el nombre: '),
            'especie': input('Ingresa la especie: '),
            'género': input('Ingresa el género: '),
            'edad en años' : input('Ingresa la edad en años: ')}

  catalogo_animales.append(registro)

  print("\n------------------------------------------")
  print("Registro creado con éxito ")
  print("------------------------------------------\n")

crear_registro()
```

```python
import matplotlib.pyplot as plt   # se importa la bibloteca para la gráfica

def ver_porcentaje(catalogo_animales):
  gatos = 0

  for animal in catalogo_animales:
  #Crear gráfico por especie
    if animal['especie'] == 'gato':
      gatos += 1

  #Definir datos y característica de la gráfica
  total_animales=len(catalogo_animales)
  valores = [gatos, total_animales-gatos]
  etiquetas = ['Gatos', 'Perros']
  colores = ['#ff9999','#66b3ff']
  explode = (0.1, 0) # Destacar la primera porción

  # 2. Crear el gráfico
  plt.pie(valores, labels=etiquetas, colors=colores, autopct='%1.1f%%',
       shadow=True, startangle=90, explode=explode)

  # 3. Personalizar y mostrar
  plt.title("Porcentaje de animales en adopción")
  plt.axis('equal')
  plt.show()

ver_porcentaje(catalogo_animales)
```

```python
# Se crea la función consultar registro

def consultar_registro(catalogo_animales):
  bregistro = False
  print("\n---------------------------------------------------------")
  nombre = input("Ingresa el nombre del animal a consultar: ")
  print("---------------------------------------------------------")
  for animal in catalogo_animales:
    if animal['nombre'] == nombre:
      bregistro = True
      indice = catalogo_animales.index(animal)
      break

  if bregistro == True:

    print(f"\nAnimal encontrado: {nombre} en el indice {indice}\n")
    print(catalogo_animales[indice])
    print("\n---------------------------------------------------------")
  else:
    print("\nAnimal no encontrado")
    print("---------------------------------------------------------")


consultar_registro(catalogo_animales)
```

```python
# Se crear la función menú para actualizar registro

def menu_actualizar():
  menu_actualizar = '''
  ==============================
  Menú actualizar registro:
  ==============================

  1. Actualizar nombre
  2. Actualizar especie
  3. Actualizar género
  4. Actualizar edad
  5. Salir del menú
  ==============================
'''
  print(menu_actualizar)


menu_actualizar()
```

```python
# Se crear función buscador registro para actualizar el registro

def consultar_registro_act(catalogo_animales):
  bregistro = False
  indice = 0
  print("\n---------------------------------------------------------")
  nombre = input("Ingresa el nombre del animal a actualizar: ")
  print("---------------------------------------------------------")

  for animal in catalogo_animales:
    if animal['nombre'] == nombre:
      bregistro = True
      indice = catalogo_animales.index(animal)+1
      break

  if bregistro == True:
    print(f"\nAnimal a actualizar encontrado: {nombre} en el indice {indice-1}\n")
    print(catalogo_animales[indice-1])
    print("---------------------------------------------------------\n")

  else:
    print("\nAnimal no encontrado")
    print("---------------------------------------------------------\n")

  return indice

consultar_registro_act(catalogo_animales)
```

```python
# Se crear función actualizar_nombre para actualizar el nombre del animal

def actualizar_nombre(catalogo_animales):

  ind=consultar_registro_act(catalogo_animales)

  if ind != 0:
    nuevo_nombre = input("Ingresa el nuevo nombre: ")
    print("---------------------------------------------------------")
    catalogo_animales[ind-1]['nombre'] = nuevo_nombre
    print("\n Registro actualizado: \n")

    print(catalogo_animales[ind-1])
    print("---------------------------------------------------------\n")

actualizar_nombre(catalogo_animales)
```

```python
# Se crear función actualizar_especie para actualizar la especie del animal

def actualizar_especie(catalogo_animales):

  ind=consultar_registro_act(catalogo_animales)

  if ind != 0:
    nueva_especie = input("Ingresa la nueva especie: ")
    print("---------------------------------------------------------")
    catalogo_animales[ind-1]['especie'] = nueva_especie
    print("\n Registro actualizado: \n")

    print(catalogo_animales[ind-1])
    print("---------------------------------------------------------\n")

actualizar_especie(catalogo_animales)
```

```python
# Se crear función actualizar_genero para actualizar el género del animal

def actualizar_genero(catalogo_animales):

  ind=consultar_registro_act(catalogo_animales)

  if ind != 0:
    nuevo_genero = input("Ingresa el nuevo género: ")
    print("---------------------------------------------------------")
    catalogo_animales[ind-1]['género'] = nuevo_genero
    print("\n Registro actualizado: \n")

    print(catalogo_animales[ind-1])
    print("---------------------------------------------------------\n")

actualizar_genero(catalogo_animales)
```

```python
# Se crear función actualizar_edad para actualizar la edad del animal

def actualizar_edad(catalogo_animales):

  ind=consultar_registro_act(catalogo_animales)

  if ind != 0:
    nueva_edad = input("Ingresa la nueva edad en años: ")
    print("---------------------------------------------------------")
    catalogo_animales[ind-1]['edad en años'] = nueva_edad
    print("\n Registro actualizado: \n")

    print(catalogo_animales[ind-1])
    print("---------------------------------------------------------\n")

actualizar_edad(catalogo_animales)
```

```python
# Se crear la función para actualizar registro

def actualizar_registro(catalogo_animales):
  while True:
    menu_actualizar()
    print("---------------------------------------------------------")
    opcion = input("Ingresa una opción: ")
    if opcion == '1':
      actualizar_nombre(catalogo_animales)
    elif opcion == '2':
      actualizar_especie(catalogo_animales)
    elif opcion == '3':
      actualizar_genero(catalogo_animales)
    elif opcion == '4':
      actualizar_edad(catalogo_animales)
    elif opcion == '5':
      print("---------------------------------------------------------")
      print("\nSaliste del menú actualizar registro")
      print("---------------------------------------------------------")
      break
    else:
      print("---------------------------------------------------------")
      print("\n Opción no encontrada")
      print("---------------------------------------------------------")



actualizar_registro(catalogo_animales)
```

```python
# Se crea una función eliminar un registro

def eliminar_registro(catalogo_animales):
  bregistro = False
  print("---------------------------------------------------------")
  nombre = input("\nIngresa el nombre del animal a eliminar: ")
  print("---------------------------------------------------------")
  for animal in catalogo_animales:
    if animal['nombre'] == nombre:
      bregistro = True
      break

  if bregistro == True:
    catalogo_animales.remove(animal)
    print("\nRegistro eliminado: \n")
    print("---------------------------------------------------------")
  else:
    print("\nRegistro no encontrado\n")
    print("---------------------------------------------------------")

  print(catalogo_animales)

eliminar_registro(catalogo_animales)
```

# **Programa principal**

```python
# Crear función menú_catalogo

def menu_catalogo():
  while True:

    menu_opciones()
    print("---------------------------------------------------------")
    opcion = input("Ingresa una opción: ")
    print("---------------------------------------------------------")
    if opcion == '1':
      ver_catalogo(catalogo_animales)
      print("\n")

  #  elif opcion == '2':
   #   ver_porcentaje(catalogo_animales)
    #  print("\n")

    elif opcion == '2':
      crear_registro()
      print("\n")

    elif opcion == '3':
      consultar_registro(catalogo_animales)
      print("\n")

    elif opcion == '4':
      actualizar_registro(catalogo_animales)
      print("\n")

    elif opcion == '5':
      eliminar_registro(catalogo_animales)
      print("\n")

    elif opcion == '6':
      print("\nSaliste del menú del catálogo")
      print("---------------------------------------------------------")
      break

    else:
      print("\nOpción no encontrada")
      print("---------------------------------------------------------")

menu_catalogo()
```
